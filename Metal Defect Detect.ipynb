```python
train_dir = "/content/dataset/NEU Metal Surface Defects Data/train"
val_dir = "/content/dataset/NEU Metal Surface Defects Data/valid"
test_dir = "/content/dataset/NEU Metal Surface Defects Data/test"

# Check dataset categories
print("Train Categories:", os.listdir(train_dir))
print("Test Categories:", os.listdir(test_dir))
print("Validation Categories:", os.listdir(val_dir))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 200  # Resize all images to 200x200 pixels
BATCH_SIZE = 32

# Data Augmentation & Normalization
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=30,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest')

val_test_datagen = ImageDataGenerator(rescale=1./255)  # Only normalization for validation/test

# Load images using generators
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(IMG_SIZE, IMG_SIZE),
                                                    batch_size=BATCH_SIZE,
                                                    class_mode='categorical')

val_generator = val_test_datagen.flow_from_directory(val_dir,
                                                     target_size=(IMG_SIZE, IMG_SIZE),
                                                     batch_size=BATCH_SIZE,
                                                     class_mode='categorical')

test_generator = val_test_datagen.flow_from_directory(test_dir,
                                                      target_size=(IMG_SIZE, IMG_SIZE),
                                                      batch_size=BATCH_SIZE,
                                                      class_mode='categorical',
                                                      shuffle=False)

# Get class labels
class_labels = list(train_generator.class_indices.keys())
print("Class Labels:", class_labels)


```
